```python
year_2011=df[df["Year"]==2011]["TotalPay"].sum()
year_2014=df[df["Year"]==2014]["TotalPay"].sum()
print("TotalPay in year 2011=",year_2011)
print("TotalPay in year 2014=",year_2014)
print("Total salary cost has increased=",year_2014-year_2011)
```
