```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
HurricanesData = pd.read_csv('Hurricanes.csv')
plt.axes([0,0,3,1.5])
plt.title('Year vs Hurricanes')
c = sns.barplot(data=HurricanesData, x='Year',y='Hurricanes')
c.set_xticklabels(c.get_xticklabels(), rotation = 90)
plt.show()
```
